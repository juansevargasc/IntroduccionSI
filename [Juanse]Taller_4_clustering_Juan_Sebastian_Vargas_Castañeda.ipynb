```
def num_elems(km, X):
    '''
    km: modelo KMeans preentrenado
    X: arreglo de numpy con los datos 

    Returns:
    num: arreglo de numpy con el número de datos por cluster
    '''
    ## Your code here
    y = km.predict(X)
    num = np.bincount(y)
    return num

```

# 4. Calcule los centroides

Escriba una función que reciba un modelo de KMeans entrenado y retorne un centroide de un cluster particular.

```
def centroid(km, n):
    '''
    km: modelo KMeans preentrenado
    n: índice del cluster

    Returns:
    centroid: arreglo de numpy con el centroide del cluster
    '''
    ## Your code here
    centroid = np.array(km.cluster_centers_)[n]
    return centroid

```

Las edades deben estar en las escala original (no escaladas entre 0 y 1).

```
def age(km, X):
    '''
    km: modelo KMeans preentrenado
    X: arreglo de numpy con los datos escalados

    Returns:
    ages: arreglo de numpy con el valor de las edades promedio
    '''
    

    ## Your code here
    y = km.predict(X)
    Xun = escaler.inverse_transform(X) # desnormaliza los datos
    ages = []
    for i in (np.unique(y)): # Recorre cada cluster
      cluage = Xun[np.argwhere(y == i), 2] #np.arg where obtiene los indices donde está cada dato del cluster y Xun[npw, 2] guarda las edades que pertenecen al cluster
      ages.append(np.mean(cluage))
    ages = np.array(ages)
    return ages

```

## 6. Calcule el porcentaje de malos deudores en cada cluster

Escriba una función que calcule el porcentaje de malos deudores en cada uno de los clusters. La función recibe un modelo preentrenado y los datos usados para entrenar el modelo y debe retornar un vector con los porcentajes.

```
def classes(km, X):
    '''
    km: modelo KMeans preentrenado
    X: arreglo de numpy con los datos 

    Returns:
    percentages: arreglo de numpy con el valor de los porcentajes
    '''
    ## Your code here
    y = km.predict(X)
    percentages=[]
    for i in (np.unique(y)): #Recorre cada cluster
      badclass = X[np.argwhere(y == i), -2] #np.arg where obtiene los indices donde está cada dato del cluster y Xun[npw, 2] guarda los malos deudores que pertenecen al cluster
      percentages.append(np.sum(badclass)/len(badclass))
    percentages= np.array(percentages)
    return percentages

```

# 8. Calcule el número de clusters usando el método del codo

Diseñe una función que reciba un modelo de kmeans (no preentrenado) y encuentre un valor *óptimo* para el número de clusters usando la estrategia del codo. El valor óptimo es el máximo valor de $k$ tal que el incremento de la inercia entre $k-1$ y $k$ es al menos de un 2%. Su función no debe crear una nueva instancia del modelo de kmeans. Debe usar la función `km.set_params(n_clusters=k)` para fijar el número de clusters.

```
def elbow(km, X, n):
    '''
    km: modelo KMeans
    X: arreglo de numpy con los datos
    n: valor máximo del número de clusters a explorar

    Returns:
    (k, inertia)
    k: número óptimo de clusters
    inertia: valor de la inercia correspondiente a k clusters
    '''
    ## Your code here
    k = 2
    arrinertia = []
    validator = True
    while (validator == True and k <= n):
      km = km.set_params(n_clusters= k)
      km = km.fit(X)
      arrinertia.append(km.inertia_)
      if (len(arrinertia) > 1):
        twopercent = arrinertia[-2]*0.02
        if (arrinertia[-2]-twopercent < arrinertia[-1]):
          validator = False
          inertia = arrinertia[-2]
      k+=1

    return (k-2, inertia)
```
