## Clase Redes Neuronales

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x): ##función de activación, se activa una vez de halla calculado los coeficientes (pesos) y sesgo. Es un último paso
    return 1.0/(1.0 + np.exp(-x))

def relu(x):
    return max(0, x) # otra función de activación

def predict(w, x): ## Predicción 
    '''
    w: numpy array of shape(9,)
    x: numpy array of shape(2,)
    '''
    a = np.zeros((3,))
    a[0] = relu(x @ w[3:5] + w[0]) # relu(x[0] * w[3] + x[1] * w[4] + w[0]) #podría ser logits x @ W + b -> b= bias o sesgo
    #print('a_0 :', a[0])
    a[1] = relu(x @ w[5:7] + w[1])
    #print('a_1 :', a[0])
    a[2] = sigmoid(a[0] * w[7] + a[1] * w[8] + w[2])
    print(a[2])
    return a[2]
```

```python
# Pruebas
w = np.array([0.1, 0.2, 0.3, -0.1, 0.5, -0.6, 0.7, 0.8, 0.9]) # 9 x 1
x = np.array([1, 0]) # 2 x 1
 
a = np.zeros((3,))
y0 = x @ w[3:5] + w[0]
a[0] = relu(y0) # relu(x[0] * w[3] + x[1] * w[4] + w[0])
print('a_0 :', a[0])
y1 = x @ w[5:7] + w[1]
a[1] = relu(y1)
print('a_1 :', a[1])
y2 = a[0] * w[7] + a[1] * w[8] + w[2]
a[2] = sigmoid(y2)
print('y2: ', y2)
```
